# Creating Copy of original data

- In Python, there are two ways to create copies
    - Shallow Copy
    - Deep Copy

Shallow Copy:- Any change made to a copy of object will we reflected in the original object as well

Deep Copy:- Any changes made to a copy of object will not be reflected in the original object

```python
cars_data1 = cars_data.copy(deep = True)
```

```python
cars_data1
```
