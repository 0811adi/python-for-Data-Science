```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars_data = pd.read_csv("C:/Users/pc/Downloads/Toyota.csv",index_col =0,na_values = ["??","????"])
```

```python
cars_data.shape
```

```python
# REMOVING ALL THE MISSING FALUE
cars_data.dropna(axis = 0,inplace = True) # axis = 0  means it removes all the row where nan values
```

```python
cars_data.shape
```

# Scatter plot

```python
# scanner plot of Price vs Age with default arguments
sns.set(style = 'darkgrid')
sns.regplot(x=cars_data['Age'],y=cars_data['Price'])

# By default fit_reg = TRrue
# It estimates and plots a regression model relating the x and y variables
```

```python
# if we not want to fit regression line
sns.regplot(x=cars_data['Age'],y=cars_data['Price'],fit_reg = False)
```

```python
sns.regplot(x=cars_data['Age'],y=cars_data['Price'],marker = "*",fit_reg = False)
```

##### Scatter  plot of Price vs Age by FuelType

```python
sns.lmplot(x = 'Age',y = 'Price',data = cars_data,
          fit_reg = False, hue = 'FuelType',
          legend = True, palette = "Set1")
```

# Histogram

```python
# Histogram with default kernal density estimate
sns.distplot(cars_data['Age'])
```

```python
# Histogram without kernal density estimate
sns.distplot(cars_data['Age'],kde = False)
```

```python
sns.displot(cars_data['Age']) # we can also get without density kernal
```

```python
# histogram with fixed no. of bins
sns.distplot(cars_data['Age'],kde = False,bins = 5)
```

# Bar plot

```python
sns.countplot(x= "FuelType",data = cars_data)
```

### Grouped bar plot

```python
# Grouped bar plot of FuelType and Automatic
sns.countplot(x= "FuelType",data = cars_data, hue = "Automatic")
```

# Box and Whiskers plot

```python
sns.boxplot(y = cars_data["Price"])
```

```python
sns.boxplot(x = cars_data["Price"])  # here price is on x axis
```

#### Box and Whiskers plot for numerical vs categorical variable

```python
# Price of cars for various fuel types
sns.boxplot(x = cars_data['FuelType'],y = cars_data["Price"])
```

#### Grouped Box and Whiskers plot for Price vs FuelType and Automatic

```python
sns.boxplot(x= 'FuelType',y = 'Price',
           hue = 'Automatic',data = cars_data)
```

# Box-Whisker plot and Histogram

```python
# lets plot Box-Whisker plot and Histogram on the same window
# first, Split the window into 2 parts


f,(ax_box,ax_hist) = plt.subplots(2,gridspec_kw={"height_ratios":(0.15,0.85)})
```

```python
f,(ax_box,ax_hist) = plt.subplots(2,gridspec_kw={"height_ratios":(0.15,0.85)})
sns.boxplot(cars_data["Price"], ax = ax_box)
sns.distplot(cars_data["Price"], ax = ax_hist, kde = False)
```

# Pairwise plots

```python
sns.pairplot(cars_data, kind = "scatter", hue = "FuelType")
plt.show()
```
