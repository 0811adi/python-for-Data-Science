# Frequency Table

pandas.crosstab()

```python
pd.crosstab(index=cars_data2['FuelType'],columns = 'count',dropna=True)
```

## Two-way table

```python
pd.crosstab(index = cars_data2['Automatic'],  # here we define 0 for manual gear box,1 for Automatic
            columns = cars_data2['FuelType'],
            dropna = True)
```

## Two-way table - joint probability

```python
pd.crosstab(index = cars_data2['Automatic'],  # here we define 0 for manual gear box,1 for Automatic
            columns = cars_data2['FuelType'],
           normalize = True,
            dropna = True)
```

## Two-way table - marginal probability

```python
pd.crosstab(index = cars_data2['Automatic'],  # here we define 0 for manual gear box,1 for Automatic
            columns = cars_data2['FuelType'],
            margins = True,
            normalize = True,
            dropna = True)
```

## Two-way table - conditional probability

```python
# given the type of gear box ,probability of different fuel type
pd.crosstab(index = cars_data2['Automatic'],  # here we define 0 for manual gear box,1 for Automatic
            columns = cars_data2['FuelType'],
            margins = True,
            normalize = 'index',
            dropna = True)
```

```python
# given the type of fuel ,probability of different gear type
pd.crosstab(index = cars_data2['Automatic'],  # here we define 0 for manual gear box,1 for Automatic
            columns = cars_data2['FuelType'],
            margins = True,
            normalize = 'columns',
            dropna = True)
```
